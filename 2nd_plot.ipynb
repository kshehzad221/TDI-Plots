```python
disputed_state_counts = disputed['State'].value_counts()
disputed_state_counts
```

### Question: Which States have the highest number of disputes?

```python
fig, ax = plt.subplots(figsize=(17,17))
ax = disputed_state_counts.plot(kind='bar', xticks=[5000,10000,15000,20000])
plt.grid(axis='y')
sns.despine(right=True, bottom=False)
```
